```python
import scipy.io as sio
import numpy as np
from copy import deepcopy
from scipy.optimize import minimize
from matplotlib import pyplot as plt
from scipy.special import erf
from mpl_toolkits.mplot3d import Axes3D
import cmath
import matplotlib
import hubbard_1d
import quspin
import fejer_kernel
import fourier_filter
import generate_cdf
from qcels_noisy_updated import *
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['lines.markersize'] = 10

num_sites = 4
J = 1.0
U = 10.0
U0 = 0.0
mu = 0.0
N_up = num_sites // 2
N_down = num_sites - N_up

num_eigenstates_max = 100

ham0 = hubbard_1d.generate_ham(num_sites, J, U0, mu, N_up, N_down)
ground_state_0 = ham0.eigsh(k=1,which="SA")[1][:,0]

ham = hubbard_1d.generate_ham(num_sites, J, U, mu, N_up, N_down)
if( num_eigenstates_max > ham.shape[0] // 2):
    eigenenergies, eigenstates = ham.eigh()
else:
    eigenenergies, eigenstates = ham.eigsh(k=num_eigenstates_max,which="SA")
ground_state = eigenstates[:,0]

population_raw = np.abs(np.dot(eigenstates.conj().T, ground_state_0))**2

# plt.plot(eigenenergies,population_raw,'b-o');plt.show()


NT = 10000
Nsample = 100
noise_alpha = 0.02
T_list = np.array([500,800,1000,2000,4000])
error_amp = np.zeros_like(T_list, dtype=float)
error_dmp = np.zeros_like(T_list, dtype=float)
# error_dmp_1 = np.zeros_like(T_list, dtype=float)
error_alpha_amp = np.zeros_like(T_list, dtype=float)
error_alpha_dmp = np.zeros_like(T_list, dtype=float)
N=5
n=0
while n<N:
    spectrum, population = generate_spectrum_population(eigenenergies, 
        population_raw, [0.7])
    ground_energy = spectrum[0]
    lambda_prior = spectrum[0]
    print(ground_energy)
    for (i,T) in enumerate(T_list):
        print("T = ", T, "n=", n+1)
        
        res, cost, max_T = \
                qcels_largeoverlap_singlelevel_noisy_linear(
                        spectrum, population, T, NT, 
                        Nsample, lambda_prior, noise_alpha,
                        fitmode='amp')
        theta = res.x[2]
        beta = np.real(res.x[3])
        error_amp[i] += np.abs(theta-ground_energy)
        error_alpha_amp[i] += np.abs(beta - noise_alpha)
        
        res, cost, max_T = \
                qcels_largeoverlap_singlelevel_noisy_linear(
                        spectrum, population, T, NT, 
                        Nsample, lambda_prior, noise_alpha,
                        fitmode='dmp')
        theta = res.x[2]
        beta = np.real(res.x[3])
        error_dmp[i] += np.abs(theta-ground_energy)
        error_alpha_dmp[i] += np.abs(beta - noise_alpha)   
    n=n+1
error_amp=error_amp/N
error_dmp=error_dmp/N
error_alpha_amp=error_alpha_amp/N
error_alpha_dmp=error_alpha_dmp/N
```

```python
print(error_dmp)
plt.figure(figsize=(12,10))
plt.plot(T_list,error_amp ,linestyle="-.",marker="o",
         label="error of new loss")
plt.plot(T_list,error_dmp,linestyle="-.",marker="*",
         label="error of old loss")
# plt.plot(T_list,error_dmp_1,linestyle="-.",marker="v",
#          label="error of old loss no beta")
plt.xlabel("$T_{max}$",fontsize=35)
plt.ylabel("error($ϵ$)",fontsize=35)
plt.xticks(fontsize=25)
plt.yticks(fontsize=25)
plt.xscale("log")
plt.yscale("log")
plt.legend(fontsize=20)
plt.savefig('Linear_increasing T.jpg')
```
